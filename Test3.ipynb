## Neglobularni klasteri

```python
import numpy as np
from GA_Clustering import GACluster
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
```

```python
n_samples = 1500
random_state = 170
X_org, y = make_blobs(n_samples=n_samples, random_state=random_state)
```

```python
plt.figure(figsize=(12, 12))
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X = np.dot(X_org, transformation)

plt.subplot(222)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

## KMeans - rezultat

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(X)
```

```python
s = silhouette_score(X, model.labels_)
s
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x=X[:,0], y=X[:, 1], c=model.labels_)
plt.title('KMeans  inertia: {} sillhouette_score: {}'.format(round(model.inertia_), round(s, 3)))
plt.show()
```

## GACluster - rezultati

```python
model = GACluster(n_clusters=3, max_iter=50, population_size=30, category='tournament', 
                  elitism_size=10, tournament_size=10, mutation_rate=0.1)
```

```python
model.fit(X)
```

```python
silhouette_score(X, model.labels_)
```

```python
model = GACluster(n_clusters=3, max_iter=20, population_size=10, 
                  elitism_size=4, mutation_rate=0.1)
```

```python
model.fit(X)
```

```python
silhouette_score(X, model.labels_)
```
